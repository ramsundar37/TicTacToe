```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
player_df = pd.read_csv("ipldata.csv")
team_df = pd.read_csv("IPLteam.csv")

# Merge on 'Team'
merged_df = pd.merge(player_df, team_df, on='Team')

# Calculate contribution % for each metric
metrics = {
    'Runs': 'Total Runs Scored',
    '4s for that team': 'Total 4s',
    '6s for that team': 'Total 6s',
    'Wickets': 'Total Wickets Taken'
}

for metric, total_metric in metrics.items():
    merged_df[f'{metric} Contribution (%)'] = (merged_df[metric] / merged_df[total_metric]) * 100
    merged_df[f'{metric} Contribution (%)'] = merged_df[f'{metric} Contribution (%)'].round(2)

# Example: Heatmap for a specific team (e.g., CSK)
team_name = 'Chennai Super Kings'  # Use full team name as in the data
team_data = merged_df[merged_df['Team'] == team_name]

# Select only the contribution columns
contrib_cols = [col for col in team_data.columns if 'Contribution (%)' in col]
heatmap_data = team_data.set_index('Player')[contrib_cols]

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, cmap='YlOrRd', fmt=".1f")
plt.title(f'{team_name} – Player Contribution % to Team Totals')
plt.ylabel('Player')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
player_df = pd.read_csv("ipldata.csv")      # Player-level stats
team_df = pd.read_csv("IPLteam.csv")        # Team-level stats

# Merge player and team data on 'Team'
merged_df = pd.merge(player_df, team_df, on='Team')

# Calculate Run Contribution %
merged_df['Run Contribution (%)'] = (merged_df['Runs'] / merged_df['Total Runs Scored']) * 100
merged_df['Run Contribution (%)'] = merged_df['Run Contribution (%)'].round(2)

# 🔍 Select a team for pie chart (e.g., CSK)
team_name = 'Chennai Super Kings'
team_players = merged_df[merged_df['Team'] == team_name]

# 🎨 Pie chart: Player run contribution to team total
plt.figure(figsize=(8, 8))
plt.pie(team_players['Run Contribution (%)'],
        labels=team_players['Player'],
        autopct='%1.1f%%',
        startangle=140,
        colors=plt.cm.tab20.colors)
plt.title(f'{team_name} Player Run Contribution to Team Total Runs')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
player_df = pd.read_csv("ipldata.csv")
team_df = pd.read_csv("IPLteam.csv")

# Clean column names
player_df.columns = player_df.columns.str.strip()
team_df.columns = team_df.columns.str.strip()

# Merge on 'Team'
merged_df = pd.merge(player_df, team_df, on='Team')

# Step 4: Rename columns after merge (optional, but keeps code consistent)
merged_df = merged_df.rename(columns={
    "Score for this team": "Runs1",
    "Total Runs Scored": "TeamRuns"
})


# Compute run contribution %
merged_df['Run Contribution (%)'] = (merged_df['Runs1'] / merged_df['TeamRuns']) * 100
merged_df['Run Contribution (%)'] = merged_df['Run Contribution (%)'].round(2)

# Select team
team_name = 'Royal Challengers Bengaluru'
team_data = merged_df[merged_df['Team'] == team_name]

# Sort and pick top 10
top_10 = team_data.sort_values(by='Run Contribution (%)', ascending=False).head(10)

# Print
print(f"\n🏏 {team_name} - Team Total Runs: {top_10['TeamRuns'].iloc[0]}") # Corrected column name
print(top_10[['Player', 'Runs1', 'TeamRuns', 'Run Contribution (%)']]) # Corrected column name

# 📈 Line Chart
plt.figure(figsize=(10, 6))
plt.plot(top_10['Player'], top_10['Run Contribution (%)'], marker='o', linestyle='-', color='purple')
plt.title(f'Top 10 {team_name} Players – Run Contribution (%)')
plt.xlabel('Player')
plt.ylabel('Run Contribution (%)')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
